## Transformada discreta de fourier

```python


# Imports
## M칩dulos para Jupyter
import warnings
warnings.filterwarnings('ignore')

import matplotlib as mpl

mpl.rcParams['figure.figsize'] = (80,30)

import matplotlib.pyplot as plt
import numpy as np
import scipy as sp

import ts2 as g
import fourier as f
```

```python
# Configuracion de los graficos y fuentes
fig_sz_x = 50
fig_sz_y = 20
fig_dpi = 200

fig_font_family = 'Ubuntu'
fig_font_size = 30
fig_label_size = 35
fig_line_width = 8
fig_grid_width = 3

def plot(tt, yy, title, xLabel = 'Tiempo [s]', yLabel = 'Amplitud'):
  
  plt.plot(tt, yy)
  plt.grid(True)
  
  # Parametros del plot
  plt.rcParams['figure.figsize'] = [fig_sz_x, fig_sz_y]  
  plt.rcParams['figure.dpi'] = fig_dpi
  plt.rcParams['lines.linewidth'] = fig_line_width
  plt.rcParams['font.family'] = fig_font_family
  plt.rcParams['font.size'] = fig_font_size
  plt.rcParams['axes.labelsize'] = fig_label_size
  plt.rcParams['grid.linewidth'] = fig_grid_width

  # Labels y titulo
  plt.title(title)
  plt.xlabel(xLabel)
  plt.ylabel(yLabel)

  plt.show()
  return

def stem(tt, xx, title, xLabel = 'Frecuencia [Hz]', yLabel = 'Amplitud'):
  
  plt.stem(tt, xx, markerfmt='C0o')
  plt.grid(True)

  # Parametros del plot
  plt.rcParams['figure.figsize'] = [fig_sz_x, fig_sz_y]  
  plt.rcParams['figure.dpi'] = fig_dpi
  plt.rcParams['lines.linewidth'] = fig_line_width
  plt.rcParams['font.family'] = fig_font_family
  plt.rcParams['font.size'] = fig_font_size
  plt.rcParams['axes.labelsize'] = fig_label_size
  plt.rcParams['grid.linewidth'] = fig_grid_width

  # Labels y titulo
  plt.title(title)
  plt.xlabel(xLabel)
  plt.ylabel(yLabel)

  plt.show()  
```

## Se침al senoidal pura
(Vemos si funciona el algoritmo)

```python
#Frecuencia de sampleo
fs = 1000

tt, xx = g.senoidal(f = 100, fs = fs)
plot(tt, xx, 'Se침al de prueba')
```

```python
XX = f.dft(xx)

# Calculo la frecuencia
N = len(XX)
nn = np.arange(N)
T = N/fs
ff = nn/T 

stem(ff, abs(XX), 'Espectro DFT')
```

```python
XX = sp.fft(xx)

# Calculo la frecuencia
N = len(XX)
nn = np.arange(N)
T = N/fs
ff = nn/T 

stem(ff, abs(XX), 'Espectro FFT')
```
