## Setup del notebook

```python
# Imports
## Módulos para Jupyter
import warnings
warnings.filterwarnings('ignore')

import matplotlib as mpl

mpl.rcParams['figure.figsize'] = (80,30)

import matplotlib.pyplot as plt

import ts1 as g
```

```python
# Configuracion de los graficos y fuentes
fig_sz_x = 50
fig_sz_y = 10
fig_dpi = 200

fig_font_family = 'Ubuntu'
fig_font_size = 30
fig_label_size = 35
fig_line_width = 5
fig_grid_width = 3

def plotter(t, y, title, x_title, y_title):
  
  # Parametros del plot
  plt.rcParams['figure.figsize'] = [fig_sz_x, fig_sz_y]  
  plt.rcParams['figure.dpi'] = fig_dpi
  plt.rcParams['lines.linewidth'] = fig_line_width
  plt.rcParams['font.family'] = fig_font_family
  plt.rcParams['font.size'] = fig_font_size
  plt.rcParams['axes.labelsize'] = fig_label_size
  plt.rcParams['grid.linewidth'] = fig_grid_width

  plt.plot(t, y)
  plt.grid(True)
  
  # Labels y titulo
  plt.title(title)
  plt.xlabel(x_title)
  plt.ylabel(y_title)

  plt.show()
```

# CODIGO DE PRUEBA

```python
tt, xx = g.senoidal(vmax = 4, dc = 3, ph = 10, ff = 100)
plotter(tt, xx, 'Senoidal', "T[s]", "Amplitud [V]")
```

### Diente de sierra

```python
tt, yy = g.diente_sierra(cc = 3, dc = 2, vmax = 10)
plotter(tt, yy, 'Diente de sierra', 'T[s]', 'Amplitud [V]')
```
